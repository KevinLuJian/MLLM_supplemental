```python
# Open a file hypo_testing/tallyqa.csv. 
# The CSV file contains a boolean column 'issimple'.
# Split the CSV into two files: tallyqa_complex and tallyqa_simple
# All rows with issimple=True go into tallyqa_simple, and all rows with issimple=False go into tallyqa_complex

import csv

with open('hypo_testing/tallyqa.csv', 'r') as csvfile:
    reader = csv.DictReader(csvfile)
    complex_rows = []
    simple_rows = []
    for row in reader:
        if row['issimple'] == 'True':
            simple_rows.append(row)
        else:
            complex_rows.append(row)

with open('hypo_testing/tallyqa_complex.csv', 'w+') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=reader.fieldnames)
    writer.writeheader()
    writer.writerows(complex_rows)

with open('hypo_testing/tallyqa_simple.csv', 'w+') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=reader.fieldnames)
    writer.writeheader()
    writer.writerows(simple_rows)
```
